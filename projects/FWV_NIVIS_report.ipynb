```python
eth_csv_path = "/Users/inesyang/Downloads/ethnicity_csv.csv"
ethnicity = pd.read_csv(eth_csv_path)

#Clean table 
ethnicity = ethnicity.rename(columns={"Column1": "Race/Ethnicity", 
    "Lifetime": "Lifetime Weighted %", 
    'Column3':'95% CI', 
    'Column4':'Lifetime Estimated # of Victims', 
    '12 Month 95% CI': '12 Month Weighted %', 
    'Column6': '12 Month 95% CI', 
    'Column7': '12 Month Estimated # of Victims'})
ethnicity = ethnicity.drop([0])
ethnicity['Race/Ethnicity'] = ethnicity['Race/Ethnicity'].replace(['Multiracial^{3}'], 'Multiracial')
ethnicity.to_csv('Ethnicity_Table8.csv', sep=',', index=False, encoding='utf-8')
ethnicity
```
